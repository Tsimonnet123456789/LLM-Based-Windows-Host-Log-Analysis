```python
# Fixed chunker
chunk_list = []

def chunker(df):
    max_chars=15000
    context=10
    start_size=80
    total_rows = len(df)
    end = 0
    chunk_count = 0
    # first chunk does not have 10 rows of context
    while end < total_rows:
        stop = min(end + start_size, total_rows)
        if end == 0:
            df_chunk = df.iloc[end:stop]
            ctx = 0
        else:
           
            df_chunk = df.iloc[max(end - context, 0):stop]
            ctx = context
            # checks to make sure there are not too many characters
        while len(df_chunk.to_string(index=False)) > max_chars and len(df_chunk) > start_size:
            df_chunk = df_chunk.iloc[:-1]  # remove last row
            stop -= 1  # removes rows if too many characters

        chunk_list.append(df_chunk)
        chunk_count += 1

        new_rows = len(df_chunk) - ctx
        if new_rows <= 0:
            new_rows = 1
        end += new_rows
        # checks to make sure the function is running and is chunking correctly
        print(f"Chunk {chunk_count}: rows {df_chunk.index[0]} to {df_chunk.index[-1]} ({len(df_chunk)} rows)")

    print(f"Total chunks created: {chunk_count}")
```

```python
# chunker(df)
print("done")
```

```python
# tests to make sure the rows are correct
all_rows = pd.concat(chunk_list, ignore_index=True)
unique_rows = all_rows['RowID'].nunique()
test1 = df['RowID'].nunique()
print(f"Unique rows: {test1}") 
print(f"Unique rows: {unique_rows}") 
```
