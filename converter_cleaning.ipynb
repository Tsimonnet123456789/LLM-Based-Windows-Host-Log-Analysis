```python
import os
import csv
from Evtx.Evtx import Evtx
import xml.etree.ElementTree as ET
# function to convert EVTX file to CSV
def Convert_evtx_to_csv(file_name):
    write = file_name.split(".")[0]
    write = os.path.join("Small_suspious_logs", write+ ".csv")
    read = os.path.join("Small_suspious_logs",file_name)
    print(read)

    # defines the XML so that the column names can be found and data
    ns = {"e": "http://schemas.microsoft.com/win/2004/08/events/event"}
    # goes through and collects column names
    with Evtx(read) as log:
        field_names = set(["RecordNumber", "Timestamp", "EventID"])
    
        for record in log.records():
            xml_data = record.xml()
            root = ET.fromstring(xml_data)
            for data in root.findall(".//e:Data", ns):
                field_names.add(data.attrib.get("Name", "Unknown"))
    
        field_names = list(field_names)  
       # writes the data to csv using the column names that were found
        with open(write, "w", newline="", encoding="utf-8") as f:
            writer = csv.DictWriter(f, fieldnames=field_names)
            writer.writeheader()
            with Evtx(read) as log2:
                for record in log2.records():
                    xml_data = record.xml()
                    root = ET.fromstring(xml_data)
                    
                    event = {
                        "RecordNumber": record.record_num(),
                        "Timestamp": root.find(".//e:TimeCreated", ns).attrib.get("SystemTime", ""),
                        "EventID": root.find(".//e:EventID", ns).text if root.find(".//e:EventID", ns) is not None else ""
                    }
    
                    for data in root.findall(".//e:Data", ns):
                        field = data.attrib.get("Name", "Unknown")
                        value = data.text if data.text else ""
                        event[field] = value
    
                    writer.writerow(event)
    
    print( write)
```
